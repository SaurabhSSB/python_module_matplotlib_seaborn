### Histogram

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel("C:\\Users\\Hp\\Desktop\\Study ITM\\Semester 2\\Applied Machine Learning\\Datasets\\drugs.xlsx")
df.head()
```

```python
plt.hist(df["Days"])
```

```python
import seaborn as sns
```

```python
sns.histplot(df["Days"])
```

```python
sns.histplot(df["Days"],kde= True)
plt.title("Investigation")
plt.xlabel("Recovery_time(days)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data= df, x= "Days",kde= True)
plt.title("Investigation")
plt.xlabel("Recovery_time(days)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(data= df, x= "Drugs",y= "Days")
plt.show()
```
